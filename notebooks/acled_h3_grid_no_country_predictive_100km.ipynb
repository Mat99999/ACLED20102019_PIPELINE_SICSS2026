# Revised H3 predictive model: 100 × 100 km grid cells, no country restriction

This Colab notebook runs the revised main-predictive H3 analysis. The original main pipeline included a country-week H3 baseline. This notebook instead matches the revised H3 definition: subsequent riots in the focal or neighboring 100 × 100 km grid cells, without restricting the outcome to the same country.

The train/test split is historical: train on 2010–2017 and test on 2018–2019.

## Model setup

- Unit: grid-cell × week
- Spatial scale: 100 × 100 km cells
- Neighborhood: focal cell plus the eight queen-neighboring cells
- Outcome: at least one riot in the local neighborhood in the next 1, 2, or 4 weeks
- Main predictor added for H3: non-peaceful protest history in focal and neighboring cells
- Models: logistic regression and random forest
- Evaluation: out-of-sample performance on 2018–2019, especially average precision
