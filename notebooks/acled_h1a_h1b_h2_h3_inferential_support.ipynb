# Inferential support tests for H1A, H1B, H2, and revised H3

This notebook runs the repository's companion inferential analysis.

The main project pipeline is predictive. These models provide supporting inferential evidence: logistic coefficients, odds ratios, confidence intervals, and p-values with two-way clustered standard errors by grid cell and week.

The revised **H3** definition removes the same-country restriction. H3 is tested as subsequent riots in the focal or neighboring 100 × 100 km grid cells, regardless of national border.

All hypotheses are tested with matched future windows:

- next 1 week
- next 2 weeks
- next 4 weeks
