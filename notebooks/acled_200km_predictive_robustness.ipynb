# ACLED predictive robustness check: 200 × 200 km grid cells

This Colab notebook reproduces the predictive robustness check for the ACLED protests and riots project. The main analysis uses 100 × 100 km grid cells. This notebook reruns the predictive analysis using larger 200 × 200 km grid cells to check whether the substantive patterns are robust to a different spatial aggregation.

The time split is unchanged: models are trained on 2010–2017 and tested on 2018–2019. The analysis is predictive rather than causal.

## What this notebook tests

- **H1A spatial diffusion:** neighboring protest histories should improve prediction of future protest risk.
- **H1B temporal diffusion:** focal-cell protest histories should improve prediction of future protest risk.
- **H2 peaceful diffusion:** peaceful protest histories should have stronger predictive signal than non-peaceful protest histories.
- **H3 protest escalation, revised:** non-peaceful protest histories should improve prediction of future riots in the focal or neighboring grid cells, without imposing a same-country restriction.

For H3, the outcome is future riot occurrence in the focal 200 × 200 km cell or the eight queen-neighboring 200 × 200 km cells over 1-, 2-, and 4-week horizons.

## H1A and H1B predictive robustness

## H2 peaceful versus non-peaceful diffusion

This table compares standardized logistic coefficients for peaceful and non-peaceful protest histories. A positive `peaceful_minus_nonpeaceful` supports H2 for that focal/neighbor and time-window contrast.

```python
h2
```

## H3 revised protest escalation robustness

This table shows the predictive gain from adding non-peaceful protest histories to a model that already includes riot history and peaceful protest history. Positive average-precision gain means the non-peaceful features improve out-of-sample prediction for that horizon/model.
